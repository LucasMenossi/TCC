```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
%matplotlib inline
import numpy as np

df20 = pd.concat(pd.read_excel("Dados/2020.xlsx", sheet_name=None), ignore_index=True)
df20 = df20.drop("Unnamed: 0", axis=1)
df19 = pd.concat(pd.read_excel("Dados/2019.xlsx", sheet_name=None), ignore_index=True)
df19 = df19.drop("Unnamed: 0", axis=1)
```

```python
df19['Placa'] = df19['Placa'].str.replace('-', '')
df20['Placa'] = df20['Placa'].str.replace('-', '')
df20
```

```python
dfresult = pd.concat([df19, df20]).reset_index()
dfresult
```

```python
df19 = df19.groupby(['Placa']).size().reset_index().rename(columns={0: 'Total'})
df20 = df20.groupby(['Placa']).size().reset_index().rename(columns={0: 'Total'})
```

```python
df19.sort_values(by='Total', ascending=False)
```

```python
df20.sort_values(by='Total', ascending=False)
```

```python
dfc = df19.merge(df20['Placa'], left_on='Placa', right_on='Placa')
dfc = dfc.drop_duplicates(subset=['Placa'])
dfc = dfresult.merge(dfc['Placa'], left_on='Placa', right_on='Placa')
dfc = dfc.drop_duplicates(subset=['Placa'])
dft = dfc.groupby(['Placa'])
dft = dft.size().reset_index()
dft
```
